```python
import cv2
print(cv2.__version__)
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(r"C:\Users\ishaan.narayan\Pictures\Screenshots\Screenshot 2025-06-26 111456.png")
```

```python
if image is None:
    print("Error: Image not found or the path is incorrect.")
else:
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 100, 200)
    contours, hierarchy = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    image_contours = image.copy()
    cv2.drawContours(image_contours, contours, -1, (0,255,0), 2)  # Draw all contours in green

    
```

```python
 # Draw all contours in green
```

```python
image_rgb = cv2.cvtColor(image_contours, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
