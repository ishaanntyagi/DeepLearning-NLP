```python
import cv2
print(cv2.__version__)
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(r"C:\Users\ishaan.narayan\Pictures\Screenshots\Screenshot 2025-06-27 100834.png")
```

```python
if image is None:
    print("Error: Image not found or the path is incorrect.")
    exit()
    
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  #conversion to grayScale 
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
