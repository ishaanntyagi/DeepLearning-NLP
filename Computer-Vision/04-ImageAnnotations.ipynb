```python
import cv2
print(cv2.__version__)
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(r"C:\Users\ishaan.narayan\Pictures\Screenshots\Screenshot 2025-06-26 111456.png")

```

```python
if image is None:
    print("Error: Image not found or the path is incorrect.")
else:
    marking = image.copy()
    # The Coordinates Here are (50,50) -> Here Represents The Top Left Corner of the Rectangle 
    # 49 -> x coordinate (Horizontal)
    # 50 -> y coordinate (vertical)
    # The coordinates here (200,200) , This Represents the Bottom Right Corner of the Rectangle 
    # 199 -> x coordinate ( horiontal position from the left )
    # 200 -> y coordinate , (Vertical position from the left )
     
    cv2.rectangle(marking, (49, 50), (199, 200), (0, 255, 0), 2)
    cv2.circle(marking, (300, 150), 40, (255, 0, 0), 2)
    cv2.line(marking, (0, 0), (400, 400), (0, 0, 255), 2)
    cv2.putText(marking, 'TEXT', (50, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)
```

```python
plt.imshow(marking)
plt.show()
```

```python
image_rgb = cv2.cvtColor(marking, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show
```
