```python
text="Artificial Intelligence and Natural Language Processing are revolutionizing the way humans interact with computers. Technologies like chatbots, voice assistants, and language translators rely on these advancements to understand and respond to human language. As AI continues to evolve, its applications in healthcare, education, and business are expanding rapidly"
```

```python
import spacy 
from nltk.stem import PorterStemmer
```

```python
nlp = spacy.load('en_core_web_sm')
doc = nlp(text)
for token in doc:
    if token.is_alpha:
        print(token.lemma)


```

```python
stemmer = PorterStemmer()
for word in text:
    a= (stemmer.stem(word))
    print(a)
```

```python
from nltk.stem import PorterStemmer

# Step 3: Create the stemmer object
stemmer = PorterStemmer()

# Step 4: List of words you want to stem
words = ["running", "runs", "runner", "easily", "fairly", "happily", "studies", "studying", "studied"]

# Step 5: Loop through each word and print its stemmed form
print("Original Word -> Stemmed Word")
for word in words:
    print(f"{word} -> {stemmer.stem(word)}")
    
```
