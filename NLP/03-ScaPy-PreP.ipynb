```python
#using spacy for tokenization
import spacy
nlp = spacy.load("en_core_web_sm")
text = "NLP is Amazing! It helps computers understand human language, even when the text is messy or unstructured."
doc = nlp(text)
for token in doc:
    tokens = token.text
    print(tokens)
```

```python
filtered_tokens = []
for token in doc:
    print(token.text)
    
print(token.is_stop)
print(token.is_punct)


        
    
    
```

```python
filtered_tokens = []

#Go through each token (word or punctuation mark) in the doc
for token in doc:
    # Check if the token is NOT a stopword (common word) and NOT punctuation
    if not token.is_stop and not token.is_punct:
        # Get the lemma (base form) and make it lowercase
        lemma = token.lemma_.lower()
        # Add it to our list
        filtered_tokens.append(lemma)

 #Print the final results
print("Filtered, lemmatized tokens:")
print(filtered_tokens)
   
```
